### Introducing the Dataset

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2021.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

### Data Cleaning

```python
# modify column names
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
# set data type at Time column
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
# sort the table with time column
exchange_rates.sort_values('Time', inplace=True)
# reset indexing
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
# 62 '-' characters
```

```python
# remove rows with '-' values, use a Boolean series
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
# set data type to float from string
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
# check what we've done
euro_to_dollar.info()
```

### Rolling Mean

```python
# import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# plot time vs USD
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.figure(figsize=(12,12))
# create the original or 1st subplot
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')
# create the other 5 plots with loop
for i, rolling_mean in zip(range(2,7),
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean()) # here calculate the rolling mean
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.loc[29:,:]
```

### Pandemic effect

We'll visualize how the euro-dollar rate has changed during the coronavirus pandemic.

```python
from_2016 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year <= 2020) ]
from_2016.sort_values(by="Time", inplace=True)
```

On 31 December 2019, WHO was informed of cases of pneumonia of unknown cause in Wuhan City, China. So we'll plot it from there until the end of 2020 (while noticing that in parts of the world it is still an ongoing problem).

```python
covid = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2020) & (euro_to_dollar['Time'].dt.year <= 2020) ]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')
# cretae plot size and select inputs, colors, width
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(from_2016['Time'], from_2016["US_dollar"].rolling(7).mean(),
        linewidth=1, color='#A6D785'
       )
ax.plot(covid['Time'], covid["US_dollar"].rolling(7).mean(),
        linewidth=2, color='#e23d28')

# Setting title
ax.text(x=16800, y=1.32, s="EUR-USD rate quickly rebounded mid pandemic", size=16, weight='bold')
ax.text(x=16800, y=1.295, s="EUR-USD exchange rate between 2016 and end of 2020", size=12)

# Highlihting Covid
ax.axvspan(xmin=18270, xmax=18625, ymin=0.085, ymax=0.98,
           alpha=0.3, color='grey')

# Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])
x = 16750
for year in ['2016', '2017', '2018', '2019', '2020']:
    ax.text(x, 1.01, year, alpha=0.6, fontsize=11)
    x += 365

y = 1.045
for rate in ['1.05', '1.1', '1.15', '1.2', '1.25']:
    ax.text(16620, y, rate, alpha=0.6, fontsize=11)
    y += 0.05

# Adding copyright
ax.text(16600, 0.983, 'Â©Zsolt Nagy' + ' '*93 + 'Source: European Central Bank',
        color = 'white', backgroundcolor = 'grey',
        size=10);
```
